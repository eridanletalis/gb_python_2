Домов без вида на набережную значительно больше, чем домов с видом.

<h3>2.2 Изучите распределение этажей домов</h3>

```python
pd.pivot_table(df, 
               values=['price_per_ft_living'],
               index='floors',
              aggfunc=['mean', 'std', np.median, 'min', 'max', 'count'])
```

```python
ff = df.floors.value_counts()
leg_lab = ['Один этаж', 
           'Один этаж + мансарда',
          'Два этажа',
          'Два этажа + мансарда',
          'Три этажа',
           'Три этажа + мансарда']
plt.figure(figsize=(10, 10))
plt.title('Распределение домов в зависимости от их этажности')

explode = [0.0, 0.1, 0.2, 0.3, 0.6, 0.8]
plt.title('Распределение домов в зависимости от их этажности')
plt.pie(ff.values,  labels=ff.index, explode=explode, )
plt.legend(leg_lab, loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title('Распределение домов в зависимости от их этажности')
plt.hist(df.floors)
plt.xlabel('Количество этажей. ".5" означает наличие мансарды')
plt.show()
```
