```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# Пример того, что следует получить
pd.get_dummies(data['whoAmI'])
```

```python
data
```

```python
index_list = data.index.values.tolist()
index_list
```

```python
uniq_values = data.whoAmI.unique()
uniq_values
```

```python
df = pd.DataFrame(columns=data.whoAmI.unique(),
                  index=range(0, len(index_list))).fillna(0)

df
```

```python
for col_name, val in data['whoAmI'].items():
	df.at[col_name, val] = 1
```
